```python
import tensorflow as tf
```

```python
from keras.src.legacy.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale =1./255,shear_range=0.2,horizontal_flip=True)
```

```python
training_set=train_datagen.flow_from_directory(r'D:\AIML-internship\car\train',
                                               target_size =(64,64),batch_size=32,class_mode='binary')
```

```python
test_datgen = ImageDataGenerator(rescale =1./255)
test_set=train_datagen.flow_from_directory(r'D:\AIML-internship\car\test',
                                           target_size =(64,64),
                                           batch_size=32,
                                           class_mode='binary')
```

```python
# part 1  convaluation
cnn = tf.keras.models.Sequential()
```

```python
# part2 add pooling
# 3 convolution
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size = 3,activation ='relu',input_shape=(64, 64, 3)))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#4 flatten
cnn.add(tf.keras.layers.Flatten())
```

```python
#5 Full connection layers
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
# output layer
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
# part-3 Training the cnn
# Assuming 'cnn' is your Keras model
cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x = training_set, validation_data = test_set,epochs = 25)
```
