```python
import tensorflow as tf
```

```python
from keras.src.legacy.preprocessing.image import ImageDataGenerator
```

# Data Preprocessing 

```python
# training set 
train_datagen = ImageDataGenerator(rescale =1./255,shear_range=0.2,horizontal_flip=True)
```

```python
training_set=train_datagen.flow_from_directory(r'D:\AIML-internship\kagglecatsanddogs_5340\train',
                                               target_size =(64,64),batch_size=32,class_mode='binary')
```

```python
# test data processing
test_datgen = ImageDataGenerator(rescale =1./255)
test_set=train_datagen.flow_from_directory(r'D:\AIML-internship\kagglecatsanddogs_5340\test',
                                               target_size =(64,64),batch_size=32,class_mode='binary')
```

```python
# part 1  convaluation
cnn = tf.keras.models.Sequential()
```

```python
# part2 add pooling
# 3 convolution
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size = 3,activation ='relu',input_shape=[64, 64, 3]))
# pooling
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

#### Adding second convolutional layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
#4 flatten
cnn.add(tf.keras.layers.Flatten())
```

```python
#5 Full connection layers
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
# output layer
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
# part-3 Training the cnn
# Assuming 'cnn' is your Keras model
cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python


cnn.fit(x = training_set, validation_data = test_set,epochs = 25)
```

```python
# part 4
import numpy as np
from keras.preprocessing import image
test_image = image.load_img(r'D:\AIML-internship\kagglecatsanddogs_5340\prediction\1.jpg',
                           target_size=(64,64))

test_image = image.img_to_array(test_image)
test_image= np.expand_dims(test_image,axis =0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
    prediction = 'Dog'
else:
    prediction ='Cat'    
```
