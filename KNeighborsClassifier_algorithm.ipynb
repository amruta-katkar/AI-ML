```python
# Import required libraries for data analysis and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the RidingMowers dataset and display it
df=pd.read_csv('RidingMowers.csv')
display(df.head())
```

```python
x=df[['Income','Lot_Size']]
y=df[['Response']]
```

```python
# Split the data into training (80%) and testing (20%) sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = KNeighborsClassifier(n_neighbors=31)
k.fit(x_train, y_train)
pre = k.predict(x_test)
```

```python
print(pre)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,pre))
```

```python
print(metrics.confusion_matrix(y_test,pre))
```

```python
print(metrics.classification_report(y_test,pre))
```
