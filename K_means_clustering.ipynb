```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
x1=df["longitude"]
x2=df["latitude"]
```

```python
from sklearn.cluster import KMeans
```

```python
c = KMeans(n_clusters=5)
```

```python
c.fit(x1.to_frame(),x2.to_frame())
```

```python
c.labels_
```

```python
c.cluster_centers_
```

```python
plt.scatter(x=df["longitude"],y=df["latitude"],c=c.labels_)
plt.show()
```
