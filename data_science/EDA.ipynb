```python
f1 ={'family':'Times new roman','color':'r','size':16}
f2 ={'family':'serif','color':'green','size':12}
f3 ={'family':'serif','color':'b','size':12}
```

```python
plt.plot(df["Glucose"],df["Age"],'+')
plt.title("Age wise glocose level",fontdict=f1)
plt.xlabel("Glucose level",fontdict=f3)
plt.ylabel("Age",fontdict=f3)
plt.show()
```

```python
plt.plot(df["Insulin"],df["Age"],'+')
plt.title("Age wise Insulin level",fontdict=f1)
plt.xlabel("Insulin level",fontdict=f3)
plt.ylabel("Age",fontdict=f3)
plt.show()
```

```python
# 1. Get the counts of each outcome
outcome_counts = df["Outcome"].value_counts()

# 2. Plot the bar chart
plt.bar(outcome_counts.index, outcome_counts.values)

# Add labels and a title for clarity
plt.xlabel("Outcome (0: No Diabetes, 1: Diabetes)")
plt.ylabel("Number of Individuals")
plt.title("Distribution of Diabetes Outcome")
plt.xticks(outcome_counts.index, labels=['No Diabetes', 'Diabetes']) # Optional: Make x-axis labels more descriptive
plt.show()
```

```python
outcome_counts = df["Outcome"].value_counts()
labels = ['No Diabetes (0)', 'Diabetes (1)']
colors = ['skyblue', 'salmon']
explode = (0.05, 0) # Slightly 'explode' the first slice

plt.pie(outcome_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow=True)
plt.title("Proportion of Diabetes Outcome")
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
