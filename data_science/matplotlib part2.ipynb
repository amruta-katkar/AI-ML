```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# histogram graph

```python
blood_sugar = [120,80,86,90,94,129,105,97,80,99]
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar,rwidth =0.8)
```

```python
plt.hist(blood_sugar,rwidth =0.4,bins=[70,90,110,130,150])
```

```python
plt.hist(blood_sugar,rwidth =0.8,bins=[70,90,110,130,150],color='g')
```

```python
blood_sugar_men = [120,80,86,90,94,129,105,97,80,99]
blood_sugar_women = [80,89,90,94,95,99,102,105,130,77]

plt.xlabel('sugar level')
plt.ylabel('no of patient')
plt.title('sugar level chart')

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,120,150],rwidth =0.95,color=['orange','yellow'],label=['men','women'])
plt.legend()
```

# piechart

```python
expence = [700,1200,300,2000,600]
expence_things = ['buss pass','food','daily travel','cloths','accessories']
```

```python
plt.pie(expence,labels=expence_things)
```

```python
plt.pie(expence,labels=expence_things)
plt.axis()
```

```python
plt.pie(expence,labels=expence_things,shadow=True)
plt.axis()
```

```python
plt.pie(expence,labels=expence_things,shadow=True,autopct='%1.1f%%')
plt.axis()
```

```python
plt.pie(expence,labels=expence_things,shadow=True,autopct='%1.1f%%',radius=1.5)
plt.axis()
```

```python
plt.pie(expence,labels=expence_things,shadow=True,autopct='%1.1f%%',radius=1.0,explode=[0.0,0.0,0.0,0.05,0.08])
plt.axis()
```

```python
plt.pie(expence,labels=expence_things,shadow=True,autopct='%1.1f%%',radius=1.0,explode=[0.0,0.0,0.0,0.05,0.08],counterclock=True)
plt.axis()
```

```python
f1 ={"family" :"Monospace","color":"r","size":15}
plt.pie(expence,labels=expence_things,shadow=True,autopct='%1.1f%%',radius=1.0,explode=[0.0,0.0,0.0,0.05,0.08],counterclock=True,startangle=45)
plt.axis('equal')
plt.title("expences",fontdict=f1)
```

```python
f1 ={"family" :"Monospace","color":"r","size":15}
plt.pie(expence,labels=expence_things,shadow=True,autopct='%1.1f%%',radius=1.0,explode=[0.0,0.0,0.0,0.05,0.08],counterclock=True,startangle=45)
plt.axis('equal')
plt.title("expences",fontdict=f1)
plt.savefig('image',bbox_inches='tight',pad_inches=1,transparent=True)  # ---> to save image/plot
```
