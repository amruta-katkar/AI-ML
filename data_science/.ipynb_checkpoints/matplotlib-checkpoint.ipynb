```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5,6])
y = np.array([100,234,123,452,456,125])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
plt.plot(x,y,'P')
plt.show()
```

```python
plt.plot(x,y,'p')
plt.show()
```

```python
plt.plot(x,y,'d')
plt.show()
```

```python
plt.plot(x,y,'s')
plt.show()
```

```python
plt.plot(x,y,'P--r')
plt.show()
```

```python
plt.plot(x,y,'p-.g')
plt.show()
```

```python
plt.plot(x,y,'s:b',ms=10)
plt.show()
```

```python
plt.plot(x,y,'s:y',ms=10,mec = 'r',color="b")
plt.show()
```

```python
plt.plot(x,y,'P',ms=10,mec='y',mfc='r',ls='-.')
plt.grid()
plt.show()
```
