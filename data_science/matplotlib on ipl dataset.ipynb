```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
player_performance_data = [
    {'player_name': 'Virat Kohli', 'runs_scored': 741, 'wickets_taken': 0},
    {'player_name': 'Jasprit Bumrah', 'runs_scored': 10, 'wickets_taken': 20},
    {'player_name': 'Andre Russell', 'runs_scored': 488, 'wickets_taken': 19},
    {'player_name': 'Ruturaj Gaikwad', 'runs_scored': 583, 'wickets_taken': 0},
    {'player_name': 'Rashid Khan', 'runs_scored': 90, 'wickets_taken': 15}
]
```

```python
overs_played = [5, 10, 15, 20]
runs_at_overs = [45, 80, 130, 180]
wickets_at_overs = [1, 2, 4, 6]
```

```python
f1 = {'family' : 'serif', 'color':'r','size':15}
f2 = {'family' : 'times new roman', 'color':'hotpink','size':12}
f3 = {'family' : 'monospace', 'color':'black','size':12}
```

```python
plt.rcParams['figure.figsize']=(10,3)

# plot 1
plt.subplot(1,3,1)
plt.plot(overs_played,wickets_at_overs,marker="*",mec="green",ms=20)
plt.title("Runs Progression",fontdict=f2)
plt.xlabel("overs_played")
plt.ylabel("wickets")

# plot 2
plt.subplot(1,3,2)
plt.plot(overs_played,wickets_at_overs,marker="*",mec="green",ms=20)
plt.title("wickets lost",fontdict=f2)
plt.xlabel("overs played")
plt.ylabel("wickets at overs")

# plot 3
plt.subplot(1,3,3)
plt.plot(overs_played,wickets_at_overs,marker="*",mec="green",ms=20)
plt.title("Runs at wickets(conceptual)",fontdict=f2)

plt.show()
```
