# view method

```python
array1 = array([100,200,300,400,23,235,6,3])
viewresult = array1.view()
array1[6] = 700
print(array1)
viewresult
```

# copy method

```python
array1 = array([100,200,300,400,23,235,6,3])
copyresult = array1.copy()
array1[5] = 800
print(array1)
copyresult
```

# reshape method

```python
a = array([1,2,3,4,5,6])

b = reshape(a,(2,3))
b
```

# flatten method

```python
flattenresult = b.flatten()
flattenresult
```
