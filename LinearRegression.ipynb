```python
# Import required libraries for data analysis and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the USA Housing dataset and display it
df=pd.read_csv('USA_Housing.csv')
display(df.head())
```

```python
# Get information about dataset structure and data types
df.info()
```

```python
# Select first 100 rows
# Feature (X): Avg. Area Number of Rooms
x=np.array(df.iloc[0:100,1]).reshape(-1,1)
```

```python
# Target (y): House Price
y=np.array(df.iloc[0:100,5]).reshape(-1,1)
```

```python
# Split the data into training (80%) and testing (20%) sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
# Create Linear Regression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
# Train the model using training data
lr.fit(x_train,y_train)
```

```python
# Predict house prices for test data
pre=lr.predict(x_test)
```

```python
# Print predicted prices
print(pre)
```

```python
# Predict price for a single input value
print(lr.predict([[5.682861]]))
```

```python
# Model intercept (bias)
lr.intercept_
```

```python
# Model coefficient (slope)
lr.coef_
```

```python
# Calculate Mean Absolute Error
from sklearn import metrics
metrics.mean_absolute_error(y_test,pre)
```

```python
# Calculate Mean Squared Error
metrics.mean_squared_error(y_test, pre)
```

```python
# Scatter plot of actual data
plt.scatter(x, y, color='blue')
plt.xlabel("Average Area Number of Rooms")
plt.ylabel("House Price")
plt.title("Linear Regression: Rooms vs Price")

# Regression line
plt.plot([min(x), max(x)], [min(pre), max(pre)], color='red')

plt.show()
```

```python
# Seaborn regression plot using dataframe
sns.regplot(
    x="Avg. Area Number of Rooms",
    y="Price",
    data=df
)
plt.title("Seaborn Regression: Rooms vs Price")
plt.show()
```

```python
sns.regplot(x="Avg. Area Number of Rooms",y= "Price",data=df)
```

```python
sns.regplot(x=x,y= y,data=df)
```

```python
# Seaborn regression plot using arrays
sns.regplot(x=x, y=y)
plt.xlabel("Average Area Number of Rooms")
plt.ylabel("House Price")
plt.title("Regression Plot Using NumPy Arrays")
plt.show()
```
